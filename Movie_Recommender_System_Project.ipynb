**Converting Unstructured Data to Structured Data**

```python
import ast 
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
  l = []
  for i in ast.literal_eval(obj):
    l.append(i['name'])
  return l
```

```python
df['genres'] = df['genres'].apply(convert)
```

```python
df.head(3)
```

```python
df['keywords'] = df['keywords'].apply(convert)
```

```python
df.head(3)
```

```python
def convert3(obj):
  l = []
  counter = 0
  for i in ast.literal_eval(obj):
    if counter != 3:
      l.append(i['name'])
      counter+=1
    else:
      break
  return l
```

```python
df['cast'] = df['cast'].apply(convert3)
```

```python
df.head(3)
```

```python
df['crew'][0]
```

**Fetching only Director Category in Job Column**

```python
def fetch_director(obj):
  l = []
  for i in ast.literal_eval(obj):
    if i["job"]=="Director":
      l.append(i['name'])
      break
  return l
```

```python
df['crew'] = df['crew'].apply(fetch_director)
```

```python
import nltk
```

**Stemming Technique**

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  l = []
  for i in text.split():
    l.append(ps.stem(i))
  return " ".join(l)  
```

**Applying stemming technique on tags column**

```python
df['tags'] = df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer 
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(df['tags']).toarray()
vectors
```

```python
cv.get_feature_names()
```

**Now, The Problem solved. No words are repeating. These are the final 5000 words from tags column.**

**Importing cosine similarity for extracting the similar words**

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
similarity[0]
```

```python
similarity[1]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x: x[1])[1:6]
```

**Creating function based on movie title to extract the similar movies**

```python
def recommend(movie):
  movie_index = df[df['title']==movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x: x[1])[1:6]
  for i in movies_list:
    print(df.iloc[i[0]].title)
```

**Recommending 5 similar movies**

```python
recommend('Batman Begins')
```

**Conclusion: This is Unsupervised Clustering problem. So, we are just finding the patterns in the data. Here, Based on the Customer input It is creating the Groups i.e. recommending similar movies.**
